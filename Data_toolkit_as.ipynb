**1. A What is NumPy, and why is it widely used in Python?**


NumPy (Numerical Python) is a powerful open-source Python library used for numerical computing. It provides efficient multi-dimensional arrays (ndarrays) that are much faster and more memory-friendly than Python lists. With built-in functions for linear algebra, statistics, random numbers, and Fourier transforms, it simplifies complex mathematical operations. NumPy also supports broadcasting, which allows operations on arrays of different shapes without explicit loops. It forms the foundation for many libraries like Pandas, Matplotlib, SciPy, scikit-learn, and TensorFlow. Because of its speed, versatility, and integration, NumPy is the backbone of scientific computing and data analysis in Python.

**2. How does broadcasting work in NumPy?**

Broadcasting in NumPy allows arithmetic operations on arrays of different shapes without copying data. It works by comparing array shapes from the trailing dimensions and stretching dimensions with size 1 to match the other. If shapes are compatible, NumPy virtually expands the smaller array during computation. This avoids loops and makes operations faster and memory-efficient. For example, adding a 1D array to a 2D array automatically aligns the smaller array across rows or columns. In short, broadcasting simplifies code and boosts performance in numerical computing.

**3. What is a Pandas DataFrame?**

A Pandas DataFrame is a two-dimensional, tabular data structure in Python, similar to an Excel sheet or SQL table. It is part of the Pandas library, which is widely used for data analysis and manipulation. A DataFrame consists of rows and columns, where each column can hold different data types (integers, floats, strings, etc.). It provides powerful features like data selection, filtering, grouping, merging, handling missing values, and statistical analysis. Because of its flexibility, ease of use, and integration with other libraries like NumPy, Matplotlib, and scikit-learn, the DataFrame is one of the most important tools for data science and machine learning in Python.

**4. A Explain the use of the groupby() method in Pandas?**


he groupby() method in Pandas is used to split data into groups based on one or more columns, so that operations like aggregation, transformation, or filtering can be performed on each group separately. It follows the Split-Apply-Combine strategy:

Apply – A function (like sum(), mean(), count(), etc.) is applied to each group.

Combine – The results are merged back into a new DataFrame.

**6. What are the differences between NumPy arrays and Python lists?**


Here are the key differences between NumPy arrays and Python lists:

Data Type – NumPy arrays store elements of the same data type, while Python lists can hold mixed data types.

Performance – NumPy arrays are faster and more memory-efficient due to optimized C-based implementation, while lists are slower.

Operations – NumPy supports element-wise mathematical operations and broadcasting, but lists require explicit loops.

Functionality – NumPy provides a wide range of mathematical, statistical, and linear algebra functions, which lists lack.

Memory – Arrays use less memory compared to lists for large datasets.

Applications – NumPy arrays are mainly used in data science, machine learning, and numerical computing, whereas lists are general-purpose.

**8. A What does the term “vectorized operation” mean in NumPy?**


In NumPy, a vectorized operation means performing operations on entire arrays or large chunks of data at once, instead of looping through individual elements. NumPy achieves this by using optimized C and Fortran implementations underneath, making computations much faster and more efficient than standard Python loops. For example, adding two arrays with a + b automatically adds corresponding elements without writing a loop. Vectorization not only improves speed and performance but also makes the code cleaner, shorter, and easier to read, which is why it is a key feature of NumPy in scientific and data-intensive computing.

**17. A Explain the difference between apply() and map() in Pandas.**


In Pandas, both **`apply()`** and **`map()`** are used to apply functions to data, but they differ in scope and usage.

* **`map()`** works only on a **Pandas Series** and applies a function **element-wise** to transform its values. It is best for simple operations like modifying or mapping individual elements.
* **`apply()`** works on both **Series and DataFrames**. On a Series, it behaves like `map()`, but on a DataFrame, it can apply a function to either **rows or columns**. This makes it more flexible for complex transformations and aggregations.

**21. Why is NumPy’s array slicing faster than Python’s list slicing?**

NumPy’s **array slicing** is faster than Python’s list slicing because NumPy arrays are stored in **contiguous blocks of memory** and implemented in optimized **C code**, while Python lists are collections of **pointers to objects** scattered in memory. When slicing, NumPy does not create a new copy of the data; instead, it returns a **view** of the original array, meaning no extra memory is used. In contrast, Python list slicing creates a **new list copy**, which takes more time and memory. This efficient memory management and vectorized backend make NumPy slicing significantly **faster and more efficient** for numerical computing.

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Math': [85, 78, 90, 95, 88],
    'Science': [80, 85, 88, 92, 86],
    'English': [78, 82, 84, 89, 90],
    'History': [70, 75, 78, 82, 85]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr = df.corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```
